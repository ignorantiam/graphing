## This is a function that will create a polynomial fit and returns coefficients of the polynomial and a coefficient of determination

```python
def get_polyfit_with_R2(x, y, degree):
    #potentially confusing functionn name, use 
    results = {}
    coeffs = np.polyfit(x, y, degree)
    
    # Polynomial Coefficients
    results['polynomial'] = coeffs.tolist()

    p = np.poly1d(coeffs)
    # fit values, and mean
    yhat = p(x)
    ybar = np.sum(y)/len(y)
    ssreg = np.sum((yhat-ybar)**2)
    sstot = np.sum((y - ybar)**2)
    # r-squared
    results['determination'] = ssreg / sstot
    return results
```

```python
x = np.array(list(final_df.Time.values), np.float)

plots = len(final_df.columns.values)
fig, axs = plt.subplots(int(math.ceil((plots-1)/2)), 2, sharex=False, sharey=False, figsize = (8,plots*1.5))
fig.subplots_adjust(hspace = .75, wspace=1.6)
#try just straight to pdf

#count for subplots
count = 0
for i in range(plots):
    #looping over dependent variables
    name = final_df.columns.values[i]
    if 'Time' in name:
        #x-axis is already Time
        pass
    else:
        #iterating over all subplots, 2 per row (r = row, c = column)
        r = math.floor(count/2)
        if count % 2 == 0:
            c = 0
        else:
            c = 1
        count += 1
        
        #Labeling subplot axes
        axs[r,c].set_xlabel("Timepoint")
        axs[r,c].set_ylabel(name)
        
        #Creating an array of values from the dependent variable
        #This is used to set the range of the full subplot, if necessary
        y = np.array(list(final_df[final_df.columns.values[i]].values), np.float)
        
        #This is a lits to seperate out values for each site
        sites = {0:['GL0', 'GL1', 'GL2', 'GL3', 'GL4'], 1:['PJ0', 'PJ1', 'PJ2', 'PJ3','PJ4'], 2:['PP0', 'PP1', 'PP2', 'PP3','PP4']}
        for k in range(len(sites)):
            #Looping through each site
            
            #creating a new dataframe of subsetted dependent variables
            sites_df = final_df.loc[sites[k]]
            
            #Setting range for y axis
            #yrange = np.amax(y) - np.amin(y)
            
            #Choosing color for each site
            color = cm(1.*k/len(sites))
            
            #creating a new list of dependent values for each site
            new_y = np.array(list(sites_df[sites_df.columns.values[i]].values), np.float)
            #In case there are missing values
            new_x = np.array(list(sites_df.Time.values), np.float)
            #Removing NaN values
            new_x = new_x[~np.isnan(new_y)] 
            new_y = new_y[~np.isnan(new_y)]
            
            #Creating continuous number line for the x range
            xs = np.linspace(np.amin(new_x), np.amax(new_x), 101)
            
            #Setting the title of the subplot
            axs[r,c].set_title(name, fontsize = 14, fontweight='bold')
            
            #Plotting (X,Y) points for each site through time
            axs[r,c].scatter(new_x, new_y, color=color)
            axs[r,c].autoscale_view()
            #Setting the x axis range
            axs[r,c].set_xlim(np.amin(x)-0.15, np.amax(x)+0.15)
            
            #Getting the equation for a 3rd order polynomial line fit
            results = get_polyfit_with_R2(new_x,new_y,3)
            #The equation for the polynomial
            p = np.poly1d(results['polynomial'])
            #Creating y values for the polynomial line
            ys = p(xs)
            
            #Creating the legend with a r-squared
            the_label = str(sites[k][0])[:-1] + ":\tR$^2$ = " + str(round(results['determination'], 2))
            #Adding the line
            axs[r,c].plot(xs, ys, color=color, label = the_label, lw = 2)
        #Setting the location of the legend
        axs[r,c].legend(bbox_to_anchor=(1, 1), loc = 2)
#Saving figure to a png
plt.savefig('Monsoon_envs_reg.png', bbox_inches='tight')
```
